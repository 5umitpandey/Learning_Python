```python
df = pd.read_excel(r"D:\SM\Python\Customer_Call_List.xlsx")
df
```

```python
#So many columns with wrong/missing/useless data
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"].str.strip()
df["Last_Name"]
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"]
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"]
```

```python
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"]
```

```python
#efficient way to do the above 3 tasks
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df["Last_Name"]
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]+', '', regex=True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('nan','', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','', regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace('Na','', regex=True)
df["Phone_Number"]
```

```python
df['Phone_Number']
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
```

```python
df
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '')
df = df.fillna('')

df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)

df
```

```python
df["Do_Not_Contact"].str.replace('', 'N')
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df

#Another way to drop null values
df = df.dropna(subset="Phone_Number"), inplace = True
```

```python
df = df.reset_index( drop=True )
df
```
