<H3> Web Scrapping With Pandas

```python
#https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue

url = r'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')[1]

#there are multiple tables in this webpage, we want the 2nd table
```

```python
table = soup.find_all('table')[1]
```

```python
soup.find_all('th')
world_titles = table.find_all('th')
world_titles

#Here we have used 'table.find_all' instead of 'soup.find_all' to only get the info about the first table headers only
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
world_table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data:
    row_data = row.find_all('td')
    print(row_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

    #We are matching the length of table headers from column_data and appneding the individual_row_data values in our panda's df
```
